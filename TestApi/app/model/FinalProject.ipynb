```python
import numpy as np
import pandas as pd  
```

```python
df = pd.read_csv('./app/model/Datasets/seattle-weather.csv')
```

```python
df
```

```python
df_processed = df
df_processed['date'] = pd.to_datetime(df_processed['date'], errors='coerce')
df_processed['month'] = df['date'].dt.month
df_processed = pd.get_dummies(df_processed, columns = ['month'], prefix = 'month', prefix_sep = '_')
```

```python
columns = list(df_processed.columns)
columns.pop(columns.index('date'))
columns.pop(columns.index('weather'))
columns.append('weather')
```

```python
df_processed = df_processed[columns]
df_processed
```

```python
X = df_processed.iloc[:, : -1].values
y = df_processed.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)
clf.fit(X_train, y_train)
```

```python
import pickle
with open("model.pkl", "wb") as f:
    pickle.dump(clf, f)
```

```python
clf.score(X_test, y_test)
```
